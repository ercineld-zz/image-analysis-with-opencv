# Image Analysis

Import necessary packages

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Display original Image

```python
img = cv2.imread('Efes.JPG')
image_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image_RGB)
plt.title('Efes')
plt.show()
```

Brighter and darker version of an image

Define Alpha and Beta for brighter version

```python
# Simple contrast control (1-3)
alpha = float(2.2)

# Simple brightness control (0-100)
beta = int(50)

result = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
```

Display brighter image

```python
image_RGB = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
plt.imshow(image_RGB)
plt.title('Efes')
plt.show()
```

Define Alpha and Beta for darker version

```python
# Simple contrast control (1-3)
alpha = float(1)

# Simple brightness control (0-100)
beta = int(-50)

result = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
```

Display darker image

```python
image_RGB = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
plt.imshow(image_RGB)
plt.title('Efes')
plt.show()
```

## Histogram Equalization

```python
img = cv2.imread('Efes.JPG',0)
equ = cv2.equalizeHist(img)
#stacking images side-by-side
res = np.hstack((img,equ))
```

Save as and display new saved image

```python
#Save as
cv2.imwrite('res.png',res)

#Display from new image
new = cv2.imread('res.png')
plt.imshow(new)
plt.title('Efes Histogram Equalized')
plt.show()
```
